```python
data = pd.read_csv('https://raw.githubusercontent.com/andreandriand/dataset/main/AirPassengers.csv')
p = data['#Passengers']
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
  X, y = list(), list()
  for i in range(len(sequence)):
    # find the end of this pattern
    end_ix = i + n_steps
    # check if we are beyond the sequence
    if end_ix > len(sequence)-1:
      break
    # gather input and output parts of the pattern
    seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
    X.append(seq_x)
    y.append(seq_y)
  return array(X), array(y)

# define univariate time series
series = p
print(series.shape)
# transform to a supervised learning problem
X, y = split_sequence(series, 2)
print(X.shape, y.shape)
# show each sample
# for i in range(len(X)):
#   print(X[i], y[i])
```

```python
newX = pd.DataFrame(X, columns=['Xt-2', 'Xt-1'])
newY = pd.DataFrame(y, columns=['Output'])
newDf = pd.concat([newX,newY], axis=1)
newDf
```
